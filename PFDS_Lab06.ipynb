3.Clean and Preprocess a Real-World Dataset (e.g., Sales Dataset) for
Analysis.

Sales Dataset â€“ Removing Duplicates

```python
import pandas as pd

# Create a sample sales dataset with duplicate entries
data = {
    'Transaction_ID': [1, 2, 3, 4, 5, 1, 6, 7, 8, 3],
    'Product': ['A', 'B', 'C', 'D', 'E', 'A', 'F', 'G', 'H', 'C'],
    'Price': [100, 200, 150, 300, 250, 100, 120, 180, 220, 150],
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-01', '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-03']
}
sales_df = pd.DataFrame(data)

print("Original DataFrame:")
display(sales_df)


```

```python
# Identify duplicates
duplicate_rows = sales_df[sales_df.duplicated()]
print("\nDuplicate Rows:")
display(duplicate_rows)

# Before duplicate count
print(f"\nNumber of rows before removing duplicates: {len(sales_df)}")

# Remove duplicates
sales_df_cleaned = sales_df.drop_duplicates()

# After duplicate count
print(f"Number of rows after removing duplicates: {len(sales_df_cleaned)}")

print("\nDataFrame after removing duplicates:")
display(sales_df_cleaned)
```

4.Clean and Preprocess a Real-World Dataset (e.g., Employee Dataset) for
Analysis.

Employee Dataset â€“ Converting Data Types

```python
import pandas as pd

# Create a sample employee dataset with incorrect data types
employee_data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': ['50000', '60000', '45000', '70000', '55000'],
    'Joining_Date': ['2022-01-15', '2021-08-20', '2023-03-10', '2020-11-01', '2022-06-25']
}
employee_df = pd.DataFrame(employee_data)

print("Original DataFrame with incorrect data types:")
display(employee_df)
print("\nOriginal data types:")
display(employee_df.dtypes)

# Convert 'Salary' to numeric
employee_df['Salary'] = pd.to_numeric(employee_df['Salary'])

# Convert 'Joining_Date' to datetime
employee_df['Joining_Date'] = pd.to_datetime(employee_df['Joining_Date'])

print("\nDataFrame after converting data types:")
display(employee_df)
print("\nCorrected data types:")
display(employee_df.dtypes)
```
